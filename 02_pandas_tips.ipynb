## 14- Splitting a string into multiple column

```python
df= pd.DataFrame({'name':['Muhammad Umair','Waqas Saeed','Asad Faiz','Hussnain Ameer'], 'location':['Karachi, Pakistan','Dubai, UAE','Lahore, Pakistan','Multan, Pakistan']})
df

```

```python
#splitting a column into two columns
df.name.str.split(' ',expand=True)
df.location.str.split(' ',expand=True)
```

```python
# adding splits into new columns
df[["first_name","last_name"]]=df.name.str.split(' ',expand=True)
df
df[["city","country"]]=df.location.str.split(', ',expand=True)
df
```

```python
# Refine columns data
df=df[['first_name','last_name','city','country']]
df
```
